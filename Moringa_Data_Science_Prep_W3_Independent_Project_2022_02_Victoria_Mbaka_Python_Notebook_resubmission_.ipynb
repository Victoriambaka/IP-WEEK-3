# Data Preparation and Data Cleaning

```python
# Preparation of cgeo by running it.
cgeo
```

```python
# There are missing values in cell_geo(cgeo), so we replace the missing values in Status, Zonename and Area
cgeo["STATUS"] = cgeo['STATUS'].fillna("disconnected")
cgeo["ZONENAME"] = cgeo['ZONENAME'].fillna("Null")
cgeo["AREA"] = cgeo['AREA'].fillna("Null")
```

```python
# Checking if cgeo information is updated.
cgeo.info()              # So it has updated the missing values.
```

cleaning the telcom datasets

```python
# Since we want to merge our Telcom datasets into one, we will need to synchronise the column names.
columns = ['Product','Value','Datetime','Cell_on_site','Number_a','Number_b','Country_A','Country_B','CELL_ID','Site_ID']
Tel_1.columns = columns
Tel_1.head()
```

```python
Tel_2.columns = columns
Tel_2.head()
```

```python

Tel_3.columns = columns
Tel_3.head()
```

```python
 Telcom = pd.concat([Tel_1, Tel_2, Tel_3], ignore_index = 1)
 Telcom
```

```python
# Dropping irrelevant columns
Clean_telcom = Telcom.drop(columns = ['Country_A' ,	'Country_B' ,'Cell_on_site'])
Clean_telcom.head()
```

```python
# Splitting date_time column into date and time
# create a new column and solit the date_time column into 2

Clean_telcom[['date', 'time']] = Clean_telcom.Datetime.str.split(expand=True)
Clean_telcom.head()
```

```python
# merge CDR and cgeo data.

cdr_cgeo = pd.merge(cgeo,Clean_telcom, how = 'inner',left_on = 'SITE_CODE', right_on = 'Site_ID')
cdr_cgeo.head(5)
```

```python
# Finding out number of most used product by grouping product and value entries.
Value = Clean_telcom.groupby(['Product'])['Value'].sum()
Value
```

* The most used product is voice at 312593 entries, data at 17287 entries, then sms at 53427 entries.

```python
# Finding out number of most used cities for the three days
Most_used_cities = cdr_cgeo.groupby(['VILLES'])['Value'].sum().head()
Most_used_cities
```

* The most used Villes are Abengourou, Abobo, Aboisso, Aboiso comoe, Abongoua
